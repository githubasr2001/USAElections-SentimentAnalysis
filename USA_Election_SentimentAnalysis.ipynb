###Sentiment Analysis

```python
from textblob import TextBlob

p=TextBlob(trump['text'][0]).sentiment.polarity
p

```

```python
def polarity(review):
  return TextBlob(review).sentiment.polarity
```

```python
trump['Sentiment Polarity']=trump['text'].apply(polarity)
biden['Sentiment Polarity']=biden['text'].apply(polarity)
```

```python
trump
```

```python
biden
```

```python
def sentiment(label):
  if label<0:
    return 'negative'
  elif label==0:
    return 'Neutral'

  else:
    return 'Positive'
```

```python
trump['Label']=trump['Sentiment Polarity'].apply(sentiment)
biden['Label']=biden['Sentiment Polarity'].apply(sentiment)
```

```python
trump
```

```python
d=trump[(trump['Label']=='Positive')|(trump['Label']=='negative')]
d
```

```python
e=biden[(biden['Label']=='Positive')|(biden['Label']=='negative')]
e
```

```python
 print(d.shape)
 print(e.shape)
```

#Sampling

```python
# Oversample df_e
from sklearn.utils import resample
e_f=resample(e,replace=True,n_samples=len(d),random_state=42)
```

```python
e_f
```

```python
biden_positive=e_f.groupby('Label').count()
b_p=(biden_positive['Sentiment Polarity'][0]/1324)*100
b_p
```

```python
trump_positive=d.groupby('Label').count()
t_p=(trump_positive['Sentiment Polarity'][0]/1324)*100
t_p
```

```python
biden_negative=e_f.groupby('Label').count()
b_n=(biden_positive['Sentiment Polarity'][1]/1324)*100
b_n
```

```python
trump_negative=d.groupby('Label').count()
t_n=(trump_positive['Sentiment Polarity'][1]/1324)*100
t_n
```

```python
import plotly.graph_objects as go
import plotly.express as xp


politicians=['Trump','Biden']
lis_pos=[t_p,b_p]
lis_neg=[t_n,b_n]

fig=go.Figure(data=[go.Bar(name='Positive',x=politicians,y=lis_pos),
                    go.Bar(name='negative',x=politicians,y=lis_neg)])

fig.update_layout(barmode='group')
fig.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report

# Combine the datasets for model training
data = pd.concat([trump[['text', 'Label']], biden[['text', 'Label']]])

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['Label'], test_size=0.2, random_state=42)

# Create a pipeline that vectorizes the text and trains a model
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# Train the model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
# Assuming the sentiment analysis and proportions calculation are done
# Let's assume these are the proportions of positive sentiments
positive_sentiment_proportion = {
    'Trump': t_p,  # Assuming t_p is the percentage of positive sentiment for Trump from your analysis
    'Biden': b_p   # Assuming b_p is the percentage of positive sentiment for Biden
}

# Simple logic to predict the winner
winner = max(positive_sentiment_proportion, key=positive_sentiment_proportion.get)

print(f"Predicted winner based on positive sentiment proportion: {winner}")
```
